```python
colors = ['Red', 'Blue', 'Green']

states = ['wa', 'nt', 'sa', 'q', 'nsw', 'v']

neighbors = {}
neighbors['wa'] = ['nt', 'sa']
neighbors['nt'] = ['wa', 'sa', 'q']
neighbors['sa'] = ['wa', 'nt', 'q', 'nsw', 'v']
neighbors['q'] = ['nt', 'sa', 'snw']
neighbors['nsw'] = ['q', 'sa', 'v']
neighbors['v'] = ['sa', 'nsw']


colors_of_states = {}

def mapping(state, color):
    for neighbor in neighbors.get(state): 
        color_of_neighbor = colors_of_states.get(neighbor)
        if color_of_neighbor == color:
            return False

    return True

def get_color_for_state(state):
    for color in colors:
        if mapping(state, color):
            return color

def main():
    for state in states:
        colors_of_states[state] = get_color_for_state(state)

    print(colors_of_states)


main()
```

```python
colors = ['Red', 'Blue', 'Green']

areas = ['WA', 'NT', 'SA', 'Q', 'NSW', 'V']

neighbors = {}
neighbors['WA'] = ['NT', 'SA']
neighbors['NT'] = ['WA', 'SA', 'Q']
neighbors['SA'] = ['WA', 'NT', 'Q', 'NSW', 'V']
neighbors['Q'] = ['NT', 'SA', 'SNW']
neighbors['NSW'] = ['Q', 'SA', 'V']
neighbors['V'] = ['SA', 'NSW']


colors_of_areas = {}

def mapping(area, color):
    for neighbor in neighbors.get(area): 
        color_of_neighbor = colors_of_areas.get(neighbor)
        if color_of_neighbor == color:
            return False
    return True

def get_color_for_area(area):
    for color in colors:
        if mapping(area, color):
            return color

def main():
    for area in areas:
        colors_of_areas[area] = get_color_for_area(area)

    print(colors_of_areas)
    for i in colors_of_areas:
      print(i,colors_of_areas[i])


main()
```
